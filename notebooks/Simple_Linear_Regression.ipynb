```python
url='https://github.com/TheEconomist/big-mac-data/raw/master/output-data/big-mac-adjusted-index.csv'
df=pd.read_csv('../data/big-mac-adjusted-index.csv', parse_dates=['date'])
# df=pd.read_csv('../data/big-mac-adjusted-index.csv')
df.head()
```

```python
df.info()
```

```python
df=df[(df['date'].dt.year==2019) & (df['date'].dt.month==1)]
df.head()
```

## sklearn: LinearRegression

```python
import sklearn
from sklearn.linear_model import LinearRegression
print(f'sklearn version: {sklearn.__version__}')
```

```python
model = LinearRegression() # create object/instance (create an instance from LinearRegression class)
model.get_params()
```

```python
df['GDP_bigmac'].head() # pandas Series
```

```python
df[['GDP_bigmac']].head()
```

```python
X = df[['GDP_bigmac']]
y = df['dollar_price']
#y = df.dollar_price
```

```python
model.fit(X, y) # step 2: fitting model
```

```python
model.score(X, y) # step 3: model scoring
```

```python
# step 4: predicting
model.predict([
    [10000],
    [20000],
    [25000]
]) # numpy array: X 2-dim array
```

```python
model.intercept_
```

```python
model.coef_
```

$y=\alpha+\beta x$

```python
model.intercept_ + model.coef_ * 9000
```

## Statsmodels: Simple Linear regression

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# formula: response ~ predictor1 + predictor2
model_a = smf.ols(formula='dollar_price ~ GDP_bigmac', data=df).fit()
```
